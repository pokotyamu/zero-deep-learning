```
import numpy as np
import matplotlib.pylab as plt

def step_function(x):
  return np.array(x > 0, dtype=np.int)  

x = np.arange(-5.0, 5.0, 0.01)
y = step_function(x)

plt.plot(x, y)
plt.show()
```

```
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

y = sigmoid(x)
plt.plot(x, y)
plt.show()
```

```
def relu(x):
  return np.maximum(0, x)

y = relu(x)
plt.plot(x, y)
plt.show()
```
